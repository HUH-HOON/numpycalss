### 데이터프레임 인덱싱

데이터 프레임을 인덱싱을 할 때도 열 라벨ㅇ르 키 값으로 생각하여 인덱싱
+ 인덱스로 라벨 값ㅇ르 하나만 넣으면 시리즈 객체가 반환되고 라벨의 배열 또는 리스트를 넣으면 부분적인 데이터 프레임이 반환  
+ 하나의 엶나 빼내면서 데이터 프레임 자료형을 유지하고 싶다면 원소가 하나인 리스트를 써서 인덱싱

### 데이터 프레임 불린 인덱싱
True, False 기반으로 인덱싱

```python
df["year"]>2014
# year가 2014보다 큰경우.
```

```python
df.loc[df['name'] == "choi"]
# 이름에서 choi 
```

```python
df.loc[df['name'] == 'choi', ['name','penalty']]
# 이름에서 choi 빼고 name. penalty 값만 가져온다.
```

```python
df.iloc[df[:, 1] == 'choi' , [1,2]]
```

```python
df['points']>2
```

```python
# numpy에서와 같이 논리연산을 응용할 수 있다.
df.loc[(df["points"]>2) & (df["points"]<3)][['name','points']] 
```

```python
a2 = df['points'] > 2
a3 = df['points'] <3
```

```python
df.loc[a2 & a3]
#여러번 쓸경우에는 a2, a3에 넣어두고 사용하기.
```
